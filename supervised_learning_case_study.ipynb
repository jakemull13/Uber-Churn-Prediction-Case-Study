## Model

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import mean_squared_error
from sklearn.model_selection import train_test_split

from sklearn.ensemble import RandomForestClassifier
```

```python
#logistic
def train_logistic_model(X, y):
    model = LogisticRegression()
    X_train, X_valid, y_train, y_valid = train_test_split(X, y, random_state=1)
    model.fit(X_train, y_train)
    #coeffs = model._coeffs
    y_predict = model.predict(X_valid)
    score = mean_squared_error(y_valid, y_predict)
    return score

#random forest
def train_random_forest(X, y, num_trees=100):
    X_train, X_valid, y_train, y_valid = train_test_split(X, y)
    rf = RandomForestClassifier(oob_score=True, n_estimators=num_trees)
    rf.fit(X_train, y_train)
    feature_importance = pd.DataFrame(rf.feature_importances_, index=X.columns)
    oob_score = rf.oob_score_
    y_predict = rf.predict(X_valid)
    print(y_predict.shape)
    print(y_valid.shape)
    score = rf.score(X_valid, y_valid)
    
    return feature_importance, score, oob_score
    
    
```

```python
feature_importance, score, oob_score = train_random_forest(X_training, y_training)
feature_importance
score
oob_score
```
